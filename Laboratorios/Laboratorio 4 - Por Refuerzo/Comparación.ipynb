```python
import numpy as np
import matplotlib.pyplot as plt

rewards_ucb = np.load("rewards_ucb.npy")
rewards_opt = np.load("rewards_optimistas.npy")
```

Vamos a realizar la comparacióon de cada método descartando el de algoritmos de gradientes debido a que este no logró un correcto aprendizaje como observamos y por lo tanto podemos descartar este método para la evaluación

```python
def smooth(y, box_size=50):
    box = np.ones(box_size) / box_size
    y_smooth = np.convolve(y, box, mode='same')
    return y_smooth
```

```python
plt.figure(figsize=(12, 6))
plt.plot(smooth(rewards_ucb), label="Intervalos de confianza (UCB)", alpha=0.8)
plt.plot(smooth(rewards_opt), label="Valores iniciales optimistas", alpha=0.8)
plt.xlabel("Episodios")
plt.ylabel("Recompensa total")
plt.title("Comparación de métodos: UCB vs. Valores optimistas")
plt.legend()
plt.grid(True)
plt.show()
```

```python
print(f"Recompensa promedio (UCB): {np.mean(rewards_ucb):.2f}")
print(f"Recompensa promedio (Optimistas): {np.mean(rewards_opt):.2f}")
```
